```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np


def plot(x, y, z, title):    
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    ax.set_title(title)
    surf = ax.plot_surface(x, y, z, alpha=0.5, color='red')
    plt.show()


def plot_cylinder(height=10, radius=3, title='3D Cylinder Plot'):

    detalization = 30
    theta = np.linspace(0, 2*np.pi, detalization)
    z = np.linspace(5, height, detalization)

    theta_grid, z_grid = np.meshgrid(theta, z)
    x_grid = radius * np.sin(theta_grid) + 10
    y_grid = radius * np.cos(theta_grid) + 10

    plot(x_grid, y_grid, z_grid, title)


def plot_sphere(radius=3, title='3D Sphere Plot'):
    d = 200

    theta = np.linspace(0, 2*np.pi, d)    
    phi = np.linspace(0, 2*np.pi, d)    

    theta_g, phi_g = np.meshgrid(theta, phi)
    
    x = radius * np.sin(theta_g)*np.cos(phi_g)
    y = radius * np.sin(theta_g)*np.sin(phi_g)
    z = radius * np.cos(theta_g)

    plot(x, y, z, title)


def plot_spiral_tube(title='3D Spiral Tube'):
    d = 120

    u = np.linspace(0, 4*np.pi, d)
    v = np.linspace(0, 2*np.pi, d)

    u, v = np.meshgrid(u, v)

    x = (2 + np.sin(v)) * np.cos(u)
    y = (2 + np.sin(v)) * np.sin(u)
    z = u + np.cos(v)

    plot(x, y, z, title)


def plot_revolution(fx, title='3D Revolution Surface'):
    d = 100
    
    theta = np.linspace(0, 2*np.pi, d)
    x = np.linspace(-5, 5, d)
    theta, x = np.meshgrid(theta, x)

    y = fx(x)*np.cos(theta)
    z = fx(x)*np.sin(theta)

    plot(x, y, z, title)

def x2(x):
    return np.sin(x) + np.cos(x)
    
```

```python
%matplotlib widget

plot_cylinder()
plot_sphere()
plot_spiral_tube()
plot_revolution(x2)
```
